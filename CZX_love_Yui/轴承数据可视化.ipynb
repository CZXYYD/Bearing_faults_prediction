```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('./Data_set/train/B01.csv')
data1 = pd.read_csv('./Data_set/train/IR01.csv')
data2 = pd.read_csv('./Data_set/train/OR01.csv')
data3 = pd.read_csv('./Data_set/train/NORMAL01.csv')
```

```python
print(data.shape)    #.shape可以看行列数

print(data.values)  #.values可以看所有值

data.values[1]   #取第一行数组
```

```python
data_t = data.T   #转置
#data_t[1]    #出来奇怪的知识
DE_time = data_t.values[1]  #取原来data的第一列数组，即为DE_time
DE_time
```

```python
data1_t = data1.T 
DE_time1 = data1_t.values[1]
data2_t = data2.T 
DE_time2 = data2_t.values[1]
data3_t = data3.T 
DE_time3= data3_t.values[1]
```

```python
t = list(range(len(DE_time)))
t1 = list(range(len(DE_time1)))
t2 = list(range(len(DE_time2)))
t3 = list(range(len(DE_time3)))
```

```python
plt.subplot(421)
plt.scatter(t,DE_time)
plt.subplot(422)
plt.scatter(t1,DE_time1,color='r')
plt.subplot(423)
plt.scatter(t2,DE_time2,color='y')
plt.subplot(424)
plt.scatter(t3,DE_time3,color='g')
```
